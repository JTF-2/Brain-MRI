```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image1 = nib.load('D:/~~Live Project -- MRI images analysis/Brain Dataset/t1/IXI102-HH-1416-T1_fcm.nii.gz')

image_t1 = image1.get_fdata()

# Putting the images in a legible orientation for us, humans.
image_t1R = np.rot90(image_t1, 2)


image2 = nib.load('D:/~~Live Project -- MRI images analysis/Brain Dataset/t2/IXI102-HH-1416-T2_reg_fcm.nii.gz')
image_t2 = image2.get_fdata()
image_t2R = np.rot90(image_t2, 2)
```

```python
# let's see a sample of what there is in the dataset

fig, axs = plt.subplots(20, 2, figsize = (50, 200))
count = 0
for slice in range(0, 100, 5):
    axs[count][0].imshow(image_t1R[slice], cmap='gray')
    axs[count][0].title.set_text("T1-w at level " +str(slice))
        
    axs[count][1].imshow(image_t2R[slice], cmap='gray')
    axs[count][1].title.set_text("T2-w at level " +str(slice))
    count+=1
        
fig.tight_layout()
        
```
